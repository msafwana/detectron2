```
import cv2
videoFile = 'video-output.mkv'
# Opens the Video file
cap= cv2.VideoCapture(videoFile)
i=0
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == False:
        break
    cv2.imwrite('Output'+str(i)+'.jpg',frame)
    i+=1
 
cap.release()
cv2.destroyAllWindows()
```
